```python
import pandas as pd
```

```python
data = pd.read_csv("SuperMarket Analysis.csv")
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
print (data.describe())
```

```python
print(data.isnull().sum())
```

```python
# converting data types to datetime format
data['Date'] = pd.to_datetime(data['Date'])
data['Time'] = pd.to_datetime(data['Time'], format='%I:%M:%S %p').dt.time
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(data, hue = 'Gender')
plt.show()
```

```python
column_name = 'Sales'

# Plot the histogram for Sales
sns.histplot(data=data, x=column_name, kde=True, color='purple')
plt.title(f'{column_name}')
plt.show()
```

```python
column_name = 'Sales'  

# Create a boxplot for Sales
plt.figure(figsize=(10, 5))
sns.boxplot(data=data, x=column_name, color='lightblue')
plt.title(f'Boxplot for {column_name}')
plt.xlabel(column_name)
plt.show()
```
